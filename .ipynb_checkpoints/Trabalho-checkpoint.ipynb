## Trabalho Extra - Unidade 01

**1.0 - Objetivo**

   O Presente trabalho tem como objetivo fazer uma análise dos dados referente aos concursos da Mega-Sena no período de março de 1996 a março de 2018. A análise está focada no comportamento dos concursos que estiveram acumulados ao longo do período estudado.

**2.0 - Metodologia**

   Os dados analisados foram obtidos a partir da planilha basedados.xlsx que contém informações de todos os concursos realizados no período citado. As plataformas utilizadas foram o Anaconda navigator e o Jupyter Notebook, e as funções desenvolvidas em linguagem Python 3.
   
   Foi necessário inserir uma coluna extra na base de dados com a informação do "Ano" em que os concursos ocorreram. Isto foi necessário pois o Notebook em uso não conseguiu converter ou formatar a coluna "DataSorteio" para que esta informasse a data no formato de ano. Mesmo usando as funções "dados.DataSorteio.dt.year" ou "pd.to_datetime(dados.DataSorteio, format= '%Y%') o resultado esperado não foi atingido.
   
   Por fim foram gerados gráficos que mostram o comportamento dos dados no período, e as análises foram feitas com base nas curvas plotadas ao longo do tempo do estudo em questão.
   
   
   **3.0 - Carga dos Dados**

A planilha basedados.xlsx carregada-se como segue: 
- Importa-se a bilioteca pandas.
- Atribui-se o nome da planilha para a variável "arquivo".
- Carrega-se a base de dados para um objeto pandas.io.excel.ExcelFile.        
- Carrega-se a planilha excel para um objeto dataframe.        
- Imprimi-se as duas primeiras linhas do dataframe.   


```python
import pandas as pd
arquivo = 'basedados.xlsx'
excel = pd.ExcelFile(arquivo)
dados = excel.parse(0)
dados.DataSorteio.dt.year

dados.head(2)
```

**4.0 - Mega-Sena Concursos Acumulados**

**4.1 - Geração de dados**

Para o estudo do comportamento da quantidade de vezes em que os concursos estiveram acumulados, a base de dados foi limitada às colunas "Ano" e "Acumulado", de modo que se pudesse contá-las ao longo dos anos através da função pivot_table.


```python
ColName = ['Ano','Acumulado']
Acumulado = dados[ColName]
Acum = dados['Acumulado']
Acumulado = Acumulado[Acum == 'SIM']
Acumulado = Acumulado.pivot_table(index = "Ano", values = "Acumulado", aggfunc="count")
```

**4.2 - Gráfico Concursos x Anos**

Foi gerado um gráfico simples de linhas utilizando-se o as funções da biblioteca "matplotlib.pyplot" para sua construção.

O comportamento do gráfico sugere um incremento considerável do número de vezes em que a Mega_Sena acumulou entre os anos de 2000 a 2006, saindo de um valor de 40 e dobrando em 2002, em seguida mantendo-se de certa forma estável no restante dos anos. Este comportamento nos levou a considerar este período a razão particular do nosso estudo, cujas considerações serão tratadas mais abaixo.



```python
import matplotlib.pyplot as plt
Acumulado.Acumulado.plot(title = 'GRÁFICO 1 - NÚMERO DE CONCURSOS ACUMULADOS POR ANO',c='blue' )
```

**4.3 - Gráfico Valores Pagos x Anos**

Igualmente foi usado um gráfico simples de linhas para sua construção. 

O comportamento do gráfico mostra um aumento gradual dos valores pagos ao longo do período podendo-se comparar, com as devidas considerações, a um comportamento linear. Esta tendência pode ser explicada pela correção monetária no valor das apostas e também pela popularização da Mega-Sena, que conduziu a um aumento no número de apostadores e ou apostas, sempre crescentes até 2015. Este fato dos dará uma base comparativa para o estudo particular do qual já mencionamos.

```python
ColName = ['Ano','Acumulado','Ganhadores_Sena','Rateio_Sena']
NaoAcumulado = dados[ColName]
Acum = NaoAcumulado['Acumulado']
NaoAcumulado = NaoAcumulado[Acum == 'NÃO']
NaoAcumulado['Premio'] = dados['Ganhadores_Sena'] * dados['Rateio_Sena']
NaoAcumulado = NaoAcumulado.pivot_table(index=['Ano'], values='Premio', aggfunc="sum")
#NaoAcumulado.head()
NaoAcumulado.Premio.plot(title='     GRÁFICO 2 - VALORES PAGOS POR ANO EM MILHÕES', c='green')
```

**4.4 - Gráfico Concursos x Ganhadores**

Este gráfico mostra comportamento do número de ganhadores da Mega-Sena ao longo dos anos e o compara com o gráfico dos concursos acumulados. O que vemos é um aumento gradual do número de ganhadores, com uma flutuação entre os anos de 2005 a 2010. Este incremento pode ser explicado pelo crescimento do número de apostas e apostadores, mas nada de extraordinário.

```python
ColName = ['Ano','Acumulado','Ganhadores_Sena']
NaoAcumulado = dados[ColName]
Acum = NaoAcumulado['Acumulado']
NaoAcumulado = NaoAcumulado[Acum == 'NÃO']
NaoAcumulado = NaoAcumulado.pivot_table(index=['Ano'], values='Ganhadores_Sena', aggfunc="count")
#NaoAcumulado.head(23)

Acumulado.Acumulado.plot(title = 'GRÁFICO 3 - NÚMERO DE CONCURSOS x GANHADORES',c='blue' )
NaoAcumulado.Ganhadores_Sena.plot(c='red')
```

**5.0 - Mega-Sena Acumulada entre os anos de 2000 a 2006**

**5.1 - Gráfico**

O gráfico abaixo detalha o crescimento do número de vezes em que a Mega-Sena acumulou no período e o compara ao gráfico de ganhadores.

**5.2 - Considerações**

1 - A princípio, somente olhando a base de dados, não podemos encontrar uma razão que justifique tal crescimento, pois em tese as regras do jogo são as mesmas até hoje.

2 - Vimos no gráfico 2 que houve crescimento dos valores pagos como prêmio, o que indica incremento da arrecadação e consequentemente de apostas, logo maior quantidade de apostas implica em menor probabilidade de não termos ganhadores. Então em tese, deveria até ter diminuido ou estabilizado a quantidade de vezes em que a Mega-Sena acumulou.

3 - Por outro lado o número de ganhadores não cresceu na mesma proporção, o que indica que no período em que esteve acumulada, a Mega-Sena juntou mais apostas do que apostadores, o que nos leva a conclusão da consideração 2.

4 - Por fim podemos perguntar por que nos anos seguintes não se observou crescimento semelhante? E podemos perguntar também por que antes do ano de 2000 o número de vezes em que a Mega-Sena acumulou era quase metade dos valores encontrados hoje? Uma vez que sabemos pelo gráfico 2 que o número de apostas era menor na época.

**6.0 - Conclusão**

Apenas com base nos dados examinados não é possível encontrar razões para explicar o comportamento dos dados estudados neste período. Entretanto o estudo serviu para indicar que respostas precisam ser dadas, e outras fontes precisam ser buscadas para completa-lo. Contudo é de conhecimento público que várias denúncias de fraudes nos sorteios da Mega-Sena tem sido feitas em vários momentos, não se deve descartaria a possibilidade.
